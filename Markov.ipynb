```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
from tqdm import tqdm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('pe_dataset.csv')
df.drop(['Unnamed: 0', 'year'], inplace=True, axis=1)
df['Hora'] = [dt.datetime.time(d) for d in pd.to_datetime(df['data'])]
df['Hora_float'] = [t.hour+t.minute/60.0 for t in df['Hora']]

df['dia'] = [int(str(d)[8:10]) for d in pd.to_datetime(df['data'])]

df_232 = df[df['br']==232]
df_104 = df[df['br']==104]

df_232.head(20)
```

```python
df_104.sort_values(by="data", inplace=True)
df_104.head(50)
```

# Em diante to fazendo só para a br_104

## Parada vai ser o seguinte.

### Gerar uma função linear qualquer com o que seria o score de risco do acidente.

100\*mortos + 20\*feridos_graves + 5\*feridos_leves + 1\*ilesos + 0\*ignorados

### Depois disso eu vou observar um treshold para esse score de forma que eu possa binarizar a parada entre: Vale a pena ir uma viatura e não vale.

```python
df_104['Score'] = 125*df_104['mortos']+25*df_104['feridos_graves']+5*df_104['feridos_leves']+1*df_104['ilesos']
```

# Para fazer o markov eu vou ter que agrupar por semana

## Juntando observações

```python
from datetime import date

df_104.data = pd.to_datetime(df_104.data, yearfirst=True, errors='ignore')

data_max = df_104.data.max()
data_min = df_104.data.min()

km_bin_size = 10
dias_aglomerados = 60

meses = (data_max - data_min).days//dias_aglomerados+2
km_bins = int((df_104.km.max() - df_104.km.min())/km_bin_size)+2
print(meses, km_bins)

matriz_dados = np.zeros((meses,km_bins))
matriz_feriados = np.zeros((meses,km_bins))
```

```python
data_start=data_min

for i in tqdm(range(matriz_dados.shape[0])):
    
    data_fim = dt.timedelta(days=dias_aglomerados) + data_start 
    km_start=0
    
    for j in range(matriz_dados.shape[1]):
        
        km_fim=km_start+km_bin_size
        
        sum_score = df_104[(df_104.km>= km_start) & (df_104.km< km_fim) & (df_104.data < data_fim) & (df_104.data >= data_start)]['Score'].sum()
        feriado_list = df_104[(df_104.km>= km_start) & (df_104.data <= data_fim) & (df_104.data>= data_start)].feriado.unique()
        
        #print(km_start, km_fim, data_start, data_fim,sum_score)
        
        if ('feriado' in feriado_list) and ('pre-feriado' in feriado_list) and ('pos-feriado' in feriado_list):
            feriado_var=1
        elif ('feriado' in feriado_list) and (('pre-feriado' in feriado_list) or ('pos-feriado' in feriado_list)):
            feriado_var=0.75
        elif ('feriado' not in feriado_list) and (('pre-feriado' in feriado_list) and ('pos-feriado' in feriado_list)):
            feriado_var=0.5
        elif ('feriado' not in feriado_list) and (('pre-feriado' in feriado_list) or ('pos-feriado' in feriado_list)):
            feriado_var=0.25
        elif ('feriado' not in feriado_list) and ('pre-feriado' not in feriado_list) and ('pos-feriado' not in feriado_list):
            feriado_var=0
            
        matriz_dados[i,j]=sum_score
        matriz_feriados[i,j]=feriado_var 
        
        km_start+=km_bin_size
        
    data_start=data_fim
```

```python
np.save('matriz_dados_br_104.npy', matriz_dados)
np.save('matriz_feriados.npy', matriz_feriados)
```

```python
def get_sum_consec_1(ndarray):
    count = 0
    for i in np.arange(1,ndarray.shape[0]):
        if ndarray[i]!=0 and ndarray[i-1]!=0:
            count+=1
    return count

def get_sum_consec_0(ndarray):
    count = 0
    for i in np.arange(1,ndarray.shape[0]):
        if ndarray[i]==0 and ndarray[i-1]==0:
            count+=1
    return count

def get_rank_value(serie_principal, serie_pesos, mes_target, verbose=False):
        
    # semana_target é a próxima semana!
    
    # Peso vai ser caculado pela correlação entre os feriados e as morte 
    # versus a distância entre a probabildiade original e 1.
    # O peso é adicionado a probabilidade original até a semana atual e anterior
    
    # Probabilidade de Morte até a semana atual
    PA_M = serie_principal[0:mes_target].sum()/(mes_target-1 + 0.001)
    D_PA_M = 1 - PA_M
    Peso_PA_M = np.sqrt(D_PA_M*np.abs(np.nan_to_num(np.corrcoef(serie_principal[0:mes_target], serie_pesos[0:mes_target])[0,1])))
    PA_M += Peso_PA_M
    
    # Probabilidade de Morte até a semana anterior
    PB_M = serie_principal[0:mes_target-1].sum()/(mes_target-2 + 0.001)
    D_PB_M = 1 - PB_M
    Peso_PB_M = np.sqrt(D_PB_M*np.abs(np.nan_to_num(np.corrcoef(serie_principal[0:mes_target-1], serie_pesos[0:mes_target-1])[0,1])))
    PB_M += Peso_PB_M
    
    # Probabilidade de ter Morte na Semana anterior, dado que teve na semana atual
    PBA_M = get_sum_consec_1(serie[0:mes_target-1])/((mes_target-2)*PB_M + 0.001) 
    
    # Probabilidade de ter Morte na semana atual, dado que teve morte na semana anterior
    PAB_M = (PBA_M*PA_M)/(PB_M + 0.001)
    
    #ANALOGAMENTE
    
    # Probabilidade de não ter morte até a semana atual
    PA_NM= 1-PA_M
    
    # Probabilidade de não ter morte até a semana anterior
    PB_NM = 1-PB_M
    
    # Probabilidade de não ter Morte na Semana anterior, dado que teve na semana atual
    PBA_NM = get_sum_consec_0(serie[0:mes_target-1])/((mes_target-2)*PB_NM+ 0.001) 
    
    # Probabilidade de não ter Morte na semana atual, dado que teve morte na semana anterior
    PAB_NM = (PBA_NM*PA_NM)/(PB_NM + 0.001)
    
    # Matriz de transição de estados A
    A = np.array([[PAB_M, 1-PAB_M], [1-PAB_NM, PAB_NM]])
    
    # Matriz para o próximo estado: previsão
    A_K = A.dot(A)
    
    # Valor do Rank é a probabilidade de ter morte, dado que não houve morte mais dado que houve morte anteriormente
    rank_value_K = (A_K[0,0]*PA_M) + (A_K[1,0]*PA_NM)
    
    debug_list = [PA_M, PA_NM, PB_M, PB_NM, PAB_M, PAB_NM, PBA_M, PBA_NM, A]
    
    return rank_value_K, debug_list
```

```python
matriz_proba = np.zeros((matriz_dados_bool.shape[0],matriz_dados_bool.shape[1]))
months_to_look = 12*2
debug = []
for km_bin in tqdm(range(matriz_proba.shape[1])):
    for month in range(matriz_proba.shape[0]):
        month_i = month - months_to_look
        if month_i<1:
            month_i = 0
        if month > 2:    
            serie = matriz_dados_bool[month_i:month-1, km_bin]
            serie_pesos = matriz_feriados[month_i:month-1, km_bin]

            matriz_proba[month, km_bin], debug_list = get_rank_value(serie, serie_pesos, month)
            debug.append(debug_list)
            
matriz_proba=np.nan_to_num(matriz_proba)
matriz_proba
```
